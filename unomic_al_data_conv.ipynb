```python
import os 
import dicom2nifti
from nipype.interfaces.dcm2nii import Dcm2niix
```

```python
dirPath = "/raid/users/usman/unomic_data/00_origData/01_P"
destDir = "/raid/users/usman/unomic_data/00_origData/01_P_Classification"
osDir = os.walk(dirPath)
osDir
```

```python
import pandas as pd
labelFile = pd.read_excel(
    "/raid/users/usman/unomic_data/00_origData/01_P.xlsx")
labelFile
```

```python
print(labelFile['Unnamed: 2'].count())
```

```python
converter = Dcm2niix()
for i in range(labelFile['Unnamed: 2'].count() - 2):

# for i in range(10): 
    patient = labelFile.iloc[i+1, 2]
    # if len(patient) 
    rpath = os.path.join(dirPath, patient)
    t1path = os.path.join(rpath, "T1")
    if os.path.isdir(rpath):
        # print(rpath)
        if os.path.isdir(t1path):
            print("Found T1", t1path)
            if not os.listdir(t1path):
                print("Empty directory")
            else:
                try:
                    converter.inputs.source_dir = t1path
                    # output_file = os.path.join(
                        # destDir, labelFile.iloc[i+1, 2] + '_' + str(int(labelFile.iloc[i+1, 6])) + '.nii')
                    if int(labelFile.iloc[i+1, 6]) == 0:
                        print("Normal class converting -- ")
                        outputFolder = os.path.join(destDir, 'Normal')
                        converter.inputs.output_dir = outputFolder
                        converter.inputs.bids_format = False
                        converter.inputs.compress = 'n'
                        # converter.inputs.out_filename = '%f'
                        converter.inputs.out_filename = patient
                        # converter
                        converter.cmdline
                        converter.run()
                        # ! dcm2niix - z n - v y - s n - o / t1path / -i / raid/users/usman/unomic_data/00_origData/01_N/01_N_DCM/
                        # print("Not empty directory")
                    elif int(labelFile.iloc[i+1, 6]) == 1:
                        print("Early MCI class converting -- ")
                        outputFolder = os.path.join(destDir, 'E_MCI')
                        converter.inputs.output_dir = outputFolder
                        converter.inputs.bids_format = False
                        converter.inputs.compress = 'n'
                        # converter.inputs.out_filename = '%f'
                        converter.inputs.out_filename = patient
                        # converter
                        converter.cmdline
                        converter.run()
                    elif int(labelFile.iloc[i+1, 6]) == 2:
                        print("Late MCI class converting -- ")
                        outputFolder = os.path.join(destDir, 'L_MCI')
                        converter.inputs.output_dir = outputFolder
                        converter.inputs.bids_format = False
                        converter.inputs.compress = 'n'
                        # converter.inputs.out_filename = '%f'
                        converter.inputs.out_filename = patient
                        # converter
                        converter.cmdline
                        converter.run()
                    elif int(labelFile.iloc[i+1, 6]) == 3:
                        print("Dementia class converting -- ")
                        outputFolder = os.path.join(destDir, 'Dementia')
                        converter.inputs.output_dir = outputFolder
                        converter.inputs.bids_format = False
                        converter.inputs.compress = 'n'
                        # converter.inputs.out_filename = '%f'
                        converter.inputs.out_filename = patient
                        # converter
                        converter.cmdline
                        converter.run()
                    else:
                        print("Class not found")
                    
                    # converter.inputs.compression = 1
                    
                except AssertionError as error:
                    print(error)
        else:
            print("T1 folder not found")
    else:
        print("Label Folder not Found")

```
